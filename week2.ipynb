```python
import pandas as pd 
import numpy as np 
from pyecharts import WordCloud
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv(r'Market_Basket_Optimisation.csv') #加载数据
```

```python
df.head() #查看数据
```

```python
df.fillna('0',inplace=True)  +空值填充为0
```

```python
df.head()
```

```python
#讲有的值append为list
marketlist=[]
count=0
for j in range(len(df.columns)):
    for i in range(len(df.index)):
        if df.iloc[i,j] != df.iloc[1,4]:
            count=count+1
            marketlist.append(df.iloc[i,j])
            print(count)
```

```python
type(marketlist)
```

```python
#输出变成list之后的值以及数量
def all_list(list1):
    result = {}
    for i in set(list1):
        result[i]=list1.count(i)
    return result
print(all_list(marketlist))
```

```python
data=all_list(marketlist)
```

```python
data.keys()
```

```python
data.values()
```

```python
#词云图
worldcloud = WordCloud(width=1300, height=620)
worldcloud.add('词云', data.keys(), data.values(), word_size_range=[20, 100])
worldcloud.render('market词云图.html')
```

```python
data1=pd.DataFrame(index=list(data.keys()))
data1['values']=data.values()
```

```python
data1
```

```python
data2=data1.sort_values(by='values',axis=0,ascending=False)
data2
```

```python
data3=data2.iloc[:10,:]
data3
```

```python
data3.index,data3.values
```

```python
plt.barh(range(len(data3)),data3['values'],tick_label = data3.index,color='rgb')
plt.show()  
```

```python
#环状图
from pyecharts import Pie
 
pie = Pie("购买最多商品前十", title_pos='center')
pie.add(
    "",
    data3.index,
    data3.values,
    radius=[40, 75],
    label_text_color=None,
    is_label_show=True,
    is_more_utils=True,
    legend_orient="vertical",
    legend_pos="left",
)
pie.render(path="购买最多商品前十环状图.html")
```
