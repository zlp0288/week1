```python
#数据可视化，简单展示前30行
#查看投诉品牌
data1=data.iloc[:30,].投诉品牌.value_counts()
```

```python
data1.index,data1.values
```

```python
#饼状图
from pyecharts import Pie
 
pie = Pie("投诉品牌统计")
pie.add(
    "",
    data1.index,
    data1.values,
    is_label_show=True,
    is_more_utils=True
)
pie.render(path="投诉品牌.html")
```

```python
from pyecharts import Bar
# 基本柱状图
bar = Bar("投诉品牌", "柱状图")
bar.use_theme('dark')  # 暗黑色主题
bar.add('',   # label
        data1.index,    # 横坐标
        data1.values,        # 纵坐标
        is_more_utils=True)    # 设置最右侧工具栏

# bar.show_config()       # 调试输出pyecharts的js的配置信息
bar.render('投诉品牌柱状图.html')  # 生成html文件
```
